```python
import pandas as pd
import matplotlib.pyplot as plt
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
import seaborn as sns

import category_encoders as ce
from sklearn import linear_model
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
pd.set_option('display.max_columns', 100)
```

```python
DATA_FOLDER_PATH = "/Users/ritumutaka/J/Categorical Feature Encoding Challenge II/csvs/cat-in-the-dat-ii"
```

```python
train = pd.read_csv(DATA_FOLDER_PATH + "/train.csv")
test = pd.read_csv(DATA_FOLDER_PATH + "/test.csv")
sample_submittion = pd.read_csv(DATA_FOLDER_PATH + "/sample_submission.csv", index_col=0)

train.sort_index(inplace=True)
train_X, train_y = train.drop(['id', 'target'], axis=1), train['target']
test_X = test.drop(['id'], axis=1)

# うまくいかないコードM
# ターゲットエンコーディング
skf = StratifiedKFold(n_splits=5, random_state=2020, shuffle=True)
# パラメータ
encode_features = list(train_X.columns)
smoothing = 0.20

#  trainデータをエンコード
oof = pd.DataFrame([])
for tr_index, oof_index in skf.split(train_X, train_y):
    ce_target_encoder = ce.TargetEncoder(cols=encode_features, smoothing=smoothing)
    ce_target_encoder.fit(train_X.iloc[tr_index, :], train_y.iloc[tr_index])
    oof_part = ce_target_encoder.transform(train_X.iloc[oof_index, :])
    oof = oof.append(oof_part, ignore_index=False)
    print(oof_index)
 
# testデータをエンコード
ce_target_encoder = ce.TargetEncoder(cols=encode_features, smoothing=smoothing)
ce_target_encoder.fit(train_X, train_y)
train_X = oof.sort_index()
test_X = ce_target_encoder.transform(test_X)

# うまくいくコード
train = pd.read_csv(DATA_FOLDER_PATH + '/train.csv')
test = pd.read_csv(DATA_FOLDER_PATH + '/test.csv')
train.sort_index(inplace=True)
train_X = train.drop(['target', 'id'], axis=1)
train_y = train['target']
test_X = test.drop(['id'], axis=1)

encode_features = list(train_X.columns)
smoothing = 0.20

skf = StratifiedKFold(n_splits=5, random_state=2020, shuffle=True)
oof = pd.DataFrame([])
for tr_idx, oof_idx in skf.split(train_X, train_y):
    ce_target_encoder = ce.TargetEncoder(cols=encode_features, smoothing=smoothing)
    ce_target_encoder.fit(train_X.iloc[tr_idx, :], train_y.iloc[tr_idx])
    oof = oof.append(ce_target_encoder.transform(train_X.iloc[oof_idx, :]), ignore_index=False)
    
ce_target_encoder = ce.TargetEncoder(cols=encode_features, smoothing=smoothing)
ce_target_encoder.fit(train_X, train_y)
train_X = oof.sort_index()
test_X = ce_target_encoder.transform(test_X)


from sklearn import linear_model
sample_submittion = pd.read_csv(DATA_FOLDER_PATH + "/sample_submission.csv", index_col=0)
glm = linear_model.LogisticRegression(
    random_state=1, solver='lbfgs', max_iter=2020, fit_intercept=True, penalty='none', verbose=0
)
glm.fit(train_X, train_y)
sample_submittion['target'] = glm.predict_proba(test_X)[:, 1]
```

```python
model_LR = linear_model.LogisticRegression(
    random_state=1, solver='lbfgs', max_iter=2020, fit_intercept=True, penalty='none', verbose=0
)
model_LR.fit(train_X, train_y)
train_y_predicted = model_LR.predict_proba(train_X)[:, 1]   # TODO: validationデータの作成
test_y_predicted = model_LR.predict_proba(test_X)[:, 1]
sample_submittion['target'] = test_y_predicted
sample_submittion.to_csv(DATA_FOLDER_PATH + "/submission_20200223.csv")
```

```python
# ターゲットの分布
def count_plot(data_series):
    values = data_series.value_counts()
    fig, ax = plt.subplots(1, 1, figsize=(8, 5))
    
    barplot = plt.bar(values.index, values, color='lightgreen', alpha=0.8)
    barplot[1].set_color('darkred')
    
    ax.set_title('Target Distribution')
    ax.annotate("percentage of target 1 : {}%".format(values.sum() / len(values)),
                xy=(0, 0), xycoords='axes fraction',
                xytext=(0, -50), textcoords='offset points',
                va="top", ha="left", color='grey',
                bbox=dict(boxstyle='round', fc="w", ec='w'))
    
    plt.xlabel('Target', fontsize=12, weight='bold')
    plt.show()
```
